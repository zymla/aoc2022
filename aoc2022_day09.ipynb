```python
with open('input_day09.txt', 'rt') as f:
    data = f.read()
data=data.splitlines()
len(data)
```

```python
h=np.array([0, 0])
t=h.copy()
t_history=[t]
for m in data:
    d, c = m.split()
    c = int(c)
    match d:
        case 'R':
            s = np.array([1, 0])
        case 'L':
            s = np.array([-1, 0])   
        case 'U':
            s = np.array([0, 1])
        case 'D':
            s = np.array([0, -1])
    for i in range(c):
        h += s
        if max(abs(h-t)) > 1:
            t = h-s
        t_history.append(t)
len(set(str(x) for x in t_history))
```
